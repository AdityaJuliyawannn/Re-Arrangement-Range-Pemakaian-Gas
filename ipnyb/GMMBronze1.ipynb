```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_excel(r"C:\Users\Aditya P J\Documents\Magang dan KKN\Magang\Proyek\data_cleanedbronze1_PGN.xlsx")
df.head()
```

```python
# Hitung rata-rata pemakaian gas per pelanggan
df_mean = df.groupby('Account Name')['Terukur m3'].mean().reset_index()
df_mean.rename(columns={'Terukur m3': 'mean_terukur_m3'}, inplace=True)

# Scaling data
scaler = StandardScaler()
df_mean['scaled_mean'] = scaler.fit_transform(df_mean[['mean_terukur_m3']])

# Fit Gaussian Mixture Model (GMM)
n_components = 2  # Tentukan jumlah cluster
gmm = GaussianMixture(n_components=n_components, random_state=42)
df_mean['gmm_cluster'] = gmm.fit_predict(df_mean[['scaled_mean']])  # Pastikan kolom 'scaled_mean' benar

# Visualisasi Hasil Clustering
plt.figure(figsize=(30, 10))
sns.scatterplot(
    x=df_mean['mean_terukur_m3'],
    y=df_mean.index,
    hue=df_mean['gmm_cluster'],
    palette='viridis',  # Ganti palette jika perlu
    s=100,             # Ukuran titik lebih besar
    alpha=0.8,
    edgecolor='k',     # Garis tepi titik
    linewidth=0.5
)

plt.title('Hasil Clustering Volume Pemakaian Gas (GMM)', fontsize=20, pad=20)
plt.xlabel('Volume Gas Terukur (m3)', fontsize=12)
plt.ylabel('Index Pelanggan', fontsize=12)

# Atur ticks dan grid
max_x = df_mean['mean_terukur_m3'].max()
plt.xticks(np.arange(0, max_x, 1000))  # Pastikan mencakup nilai maksimum
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Tambahkan legend
plt.legend(
    title='Cluster', 
    bbox_to_anchor=(1.05, 1),  # Posisi legend di luar plot
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

```python
cluster_stats = df_mean.groupby('gmm_cluster')['mean_terukur_m3'].describe()

print('\nDeskripsi Statistik Volume Pemakaian per Cluster :')
print(cluster_stats.to_markdown(tablefmt='grid', floatfmt='.2f'))
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(
    data=df_mean,
    x='mean_terukur_m3',  # Gunakan kolom rata-rata yang sudah dihitung
    hue='gmm_cluster',    # Gunakan kolom hasil clustering GMM
    bins=30,              # Jumlah bins disesuaikan (100 terlalu banyak untuk distribusi gas)
    palette='Set2',
    multiple='stack',     # Tumpuk histogram per cluster
    kde=True              # Tambahkan garis density estimate (opsional)
)

plt.title('Distribusi Volume Gas per Klaster (GMM)', fontsize=14)
plt.xlabel('Volume Gas (m3)', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)

# Atur legend
plt.legend(
    title='Cluster',
    labels=[f'Cluster {i}' for i in range(n_components)],  # Misal: Cluster 0, 1, 2
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

```python
summary = df_mean.groupby('gmm_cluster')['mean_terukur_m3'].agg(['count', 'min', 'max', 'mean', 'std']).sort_values('min')
summary['range'] = summary['min'].astype(int).astype(str) + " - " + summary['max'].astype(int).astype(str)
print(summary)
```

```python
persen = df_mean['gmm_cluster'].value_counts(normalize=True).sort_index() * 100
print("Persentase pelanggan per klaster:\n", persen)
```

```python
plt.figure(figsize=(12, 6))

# Plot density untuk setiap cluster
for i in range(n_components):
    sns.kdeplot(df_mean[df_mean['gmm_cluster'] == i]['mean_terukur_m3'], 
                label=f'Cluster {i}', 
                shade=True, 
                alpha=0.6)

plt.title('Density Plot of GMM Clusters')
plt.xlabel('Mean Terukur m3')
plt.ylabel('Density')
plt.legend(title='Clusters')
plt.tight_layout()
plt.show()
```

```python
probs = gmm.predict_proba(df_mean[['scaled_mean']])

plt.figure(figsize=(20, 6))
plt.scatter(df_mean['mean_terukur_m3'], 
            df_mean.index, 
            c=probs.max(axis=1), 
            cmap='viridis', 
            s=50, alpha=0.7)
plt.colorbar(label='Max Probability')
plt.title('Clustering Probabilities for GMM')
plt.xlabel('Mean Terukur m3')
max_x = df_mean['mean_terukur_m3'].max()
plt.xticks(np.arange(0, max_x, 1000), fontsize = 14) 
plt.yticks(fontsize = 14)
plt.ylabel('Index Pelanggan')
plt.tight_layout()
plt.show()
```
