```python
df['Periode'] = pd.to_datetime(df['Periode'], errors='coerce')
df.describe()
```

```python
def normalisasi_data (batas_min, batas_max):
    if 1000 <= batas_min <= 1601 and 6000 <= batas_max <= 16500:
        segmentasi = 'Bronze 1'
    elif batas_min <= 350 and batas_max <= 4000:
        segmentasi = 'Bronze 2'
    else:
        segmentasi = 'Tidak Valid'

    return segmentasi
```

```python
df['segmentasi'] = df.apply(
    lambda row: normalisasi_data(row['Sum of Batas Min'], row['Sum of Batas Max']), axis = 1)
df.head()
```
